```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Delivery wise Analysis

```python
delivery = pd.read_csv("Indian Premier League/deliveries.csv")
delivery.head(10)
```

```python
print(delivery.shape)
print(delivery.columns)
```

```python
delivery.describe()
```

```python
delivery.info()
```

```python
l = list(delivery["dismissal_kind"].unique())[1:]
data = {}
for i in l:
    data[i] = delivery[delivery["dismissal_kind"]==i].shape[0]
df = pd.Series(data)
df = df.reset_index()
df.columns=["dismissal_kinds","count"]
df
```

```python
import plotly.express as px
fig = px.pie(df, values='count', names='dismissal_kinds', title='% Dismissls by Dismissal kinds')
fig.show()
```

```python
top_scorers = delivery.groupby('batsman').sum()['batsman_runs'].sort_values(ascending=False).head(10)
top_scorers = top_scorers.reset_index()
top_scorers.columns= ["Batsman", "No of runs scored"]
plt.figure(figsize=(15,8))
sns.barplot(data= top_scorers, x="Batsman", y="No of runs scored")
plt.ylabel("No of runs scored", fontsize=20)
plt.xlabel("Batsman", fontsize=20)
plt.title("Top 10 highest run scorers", fontsize=30)
plt.xticks(rotation=90)
plt.show()
```

```python
sixes=delivery[delivery['batsman_runs']==6]
sixes = sixes.groupby('batsman').sum()['batsman_runs'].sort_values(ascending=False).head(10)
sixes = sixes.reset_index()
sixes.columns= ["Batsman", "sixes"]
plt.figure(figsize=(15,8))
sns.barplot(data= sixes, x="Batsman", y="sixes")
plt.ylabel("No of sixes scored", fontsize=20)
plt.xlabel("Batsman", fontsize=20)
plt.title("Top 10 highest six scorers", fontsize=30)
plt.xticks(rotation=90)
plt.show()
```

```python
six = delivery[delivery['batsman_runs']==6]
pt=six.pivot_table(index='over',columns='batting_team',values='batsman_runs', aggfunc='count')
plt.figure(figsize=(15,8))
sns.heatmap(pt)
plt.xlabel("Team", fontsize=20)
plt.title("Teamwise sixes scored in each_over", fontsize=30)
plt.ylabel("Over Number", fontsize=20)
plt.show()
```

```python
fours=delivery[delivery['batsman_runs']==4]
fours = fours.groupby('batsman').sum()['batsman_runs'].sort_values(ascending=False).head(10)
fours = fours.reset_index()
fours.columns= ["Batsman", "four"]
plt.figure(figsize=(15,8))
sns.barplot(data= fours, x="Batsman", y="four")
plt.ylabel("No of fours scored", fontsize=20)
plt.xlabel("Batsman", fontsize=20)
plt.title("Top 10 highest four scorers", fontsize=30)
plt.xticks(rotation=90)
plt.show()
```

```python
four = delivery[delivery['batsman_runs']==6]
pt=four.pivot_table(index='over',columns='batting_team',values='batsman_runs', aggfunc='count')
plt.figure(figsize=(15,8))
sns.heatmap(pt)
plt.xlabel("Team", fontsize=20)
plt.title("Teamwise fours scored in each_over", fontsize=30)
plt.ylabel("Over Number", fontsize=20)
plt.show()
```

```python
from ipywidgets import interact
@interact
def delivery_wise_boundary(ball=list(delivery["ball"].unique())):
    a=delivery[(delivery['ball']==ball)]
    a=pd.DataFrame(a['batsman_runs'].value_counts())
    a = a.rename(columns={'batsman_runs':'runs_scored'})
    return a.style.background_gradient(cmap='plasma')
```

```python
@interact
def over_wise_scores(over=list(delivery["over"].unique())):
    a=delivery[(delivery['over']==over)]
    a=pd.DataFrame(a['batsman_runs'].value_counts())
    a = a.rename(columns={'batsman_runs':'runs_scored'})
    return a.style.background_gradient(cmap='coolwarm')
```

```python
matches["toss+match_win"] = (matches["toss_winner"]==matches["winner"]).astype("int")
```

```python
plt.figure(figsize=(15,8))
sns.countplot(data=matches, x="toss_decision",  hue = "toss+match_win", palette = 'plasma')
plt.title("matches won or lost after winning the toss and electiong to bat or ball", fontsize = 20)
plt.xlabel("Decision after winning the toss",fontsize=15)
plt.ylabel("No of matches", fontsize=15)
plt.show()
```

```python
@interact
def match_city(season=list(matches['season'].unique())):
    a=matches[(matches['season']==season)]
    a=pd.DataFrame(a['city'].value_counts())
    return a.style.background_gradient(cmap='plasma')
```

```python
@interact
def match_won_seasonwise(season=list(matches['season'].unique())):
    a=matches[(matches['season']==season)]
    a=pd.DataFrame(a['winner'].value_counts())
    a = a.rename(columns={"winner":"Matches won"})
    return a.style.background_gradient(cmap='plasma')
```

## Analysis on Merged Data

```python
merged=delivery.merge(matches,left_on='match_id',right_on='id')
merged.head()
```

```python
print(merged.columns)
```

```python
def plotCareerCurve(batsman):
    return merged[merged['batsman']==batsman].groupby('season').sum()['batsman_runs'].plot(kind='line')
```
